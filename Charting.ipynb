```python
import random
import statistics
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def flip_coin():
    for _ in range(1):
        flip = random.random()
        if flip > 0.5:
            return 'H'
        else:
            return 'T'
        
```

```python
def simulation(n = 16):
    count = 0
    head_count = 0
    tail_count = 0
    heads = []
    tails = []
    while count < 2 ** n:
        flip = flip_coin()
        if flip == "H":
            head_count += 1
        elif flip == "T":
            tail_count += 1
        
        count += 1
        for i in range(n + 1):
            if count == 2 ** i:
                heads.append(head_count)
                tails.append(tail_count)
    return heads, tails
            
            
```

```python
H, T = simulation()
```

```python
print(H, T)
```

```python
diff = [H[i] - T[i] for i in range(len(H))]
```

```python
print(diff)
```

```python
y1 = diff
plt.plot(y1, color="green")
plt.xlabel("2 ** n intervals")
plt.ylabel("Difference")
plt.title("Difference between heads and tails")
plt.show()
```

```python
def ratio(a, b):
    ratio = []
    for i in range(len(a)):
        if b[i] != 0:
            ratio.append(a[i]/b[i])
        else:
            ratio.append(0)
    return ratio
```

```python
ratio = ratio(H, T)
```

```python
print(ratio)
```

```python
y1 = ratio
plt.plot(y1, color="green")
plt.xlabel("2 ** n intervals")
plt.ylabel("Ratio")
plt.title("Ratio of heads to tails")
plt.show()
```

```python
def sim_20():
    ratio20 = []
    for i in range(20):
        my_list = []
        H, T = simulation()
        for j in range(len(H)):
            if T[j] != 0:
                my_list.append(H[j] / T[j])
            else:
                my_list.append(0)
        ratio20.append(my_list)
    return ratio20

ratio20 = sim_20()
```

```python
def find_mean(x):
    mean = []
    for i in range(len(x[0])):
        my_list = []
        for j in range(len(x)):
            my_list.append(x[j][i])
        mean.append(statistics.mean(my_list))
    return mean
        
    
```

```python
mean = find_mean(ratio20)
```

```python
print(mean)
```

```python
y1 = mean
x1 = [2**n for n in range(17)]
plt.scatter(x1,y1)
plt.xscale('log')
plt.xlabel("")
plt.ylabel("Mean")
plt.title("Mean of Head/Tails ratio")
plt.show()
```

```python
def find_stdev(x):
    stdev = []
    for i in range(len(x[0])):
        my_list = []
        for j in range(len(x)):
            my_list.append(x[j][i])
        stdev.append(statistics.stdev(my_list))
    return stdev
```

```python
stdev = find_stdev(ratio20)
```

```python
print(stdev)
```

```python
y1 = stdev
x1 = [2**n for n in range(17)]
plt.scatter(x1,y1)
plt.xscale('log')
plt.xlabel("")
plt.ylabel("Standandard Deviation")
plt.title("St. Deviation of Head/Tails ratio")
plt.show()
```
